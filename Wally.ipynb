```python
import numpy as np
import cv2 as cv

# Cargar los clasificadores
#face_cascade = cv.CascadeClassifier('haarcascade_frontalface_alt2.xml')
wally_cascade = cv.CascadeClassifier('cascade2.xml')

# Leer la imagen
frame = cv.imread("C:\\Users\\kevin\\Desktop\\wally11.png")

# Convertir la imagen a escala de grises
gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

# Detectar a Wally usando el clasificador en cascada
wallys = wally_cascade.detectMultiScale(gray, 1.3, 5)



# Dibujar rectángulos alrededor de las detecciones
for (x, y, w, h) in wallys:
    frame = cv.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Mostrar la imagen con las detecciones
cv.imshow('Deteccion', frame)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import os

# Ruta de la carpeta con las imágenes originales
folder_path = 'C:\\Users\\kevin\\Desktop\\prueba\\'

# Ruta de la carpeta para guardar las imágenes redimensionadas
output_folder = 'C:\\Users\\kevin\\Desktop\\prueba2\\'

# Obtener lista de nombres de archivos en la carpeta
image_files = os.listdir(folder_path)

# Contador para el nombre de las imágenes redimensionadas
count = 1

# Iterar sobre cada imagen en la carpeta
for filename in image_files:
    # Leer la imagen
    img = cv.imread(os.path.join(folder_path, filename), 1)
    
    # Redimensionar la imagen a 50x50
    img_resized = cv.resize(img, (50, 50), interpolation=cv.INTER_AREA)
    
    # Guardar la imagen redimensionada en la carpeta de salida
    output_path = os.path.join(output_folder, f'wallyp{count}.jpg')
    cv.imwrite(output_path, img_resized)

    # Incrementar el contador
    count += 1

print('Proceso completado. Imágenes redimensionadas y guardadas en la carpeta "p".')
```
